```python
cell_type_row = pd.DataFrame({'Unnamed: 0':['cellType']})
e7_t01_rawCount_data = pd.concat([cell_type_row, e7_t01_rawCount_data]).reset_index(drop = True)
e7_t02_rawCount_data = pd.concat([cell_type_row, e7_t02_rawCount_data]).reset_index(drop = True)
e7_t01_rawCount_data.set_index('Unnamed: 0')
e7_t02_rawCount_data.set_index('Unnamed: 0')
```

```python
e7_t01_rawCount_data['mean'] = e7_t01_rawCount_data.mean(axis=1) #assign mean value to each row in dataframe
e7_t02_rawCount_data['mean'] = e7_t02_rawCount_data.mean(axis=1)
```

```python
e7_t01_rawCount_data.replace(np.NaN, ' ', inplace=True)
e7_t02_rawCount_data.replace(np.NaN, ' ', inplace=True)
```

```python
cell_types_list = ['Alpha','Beta','Delta','Gamma','Epsilon'] #the cell types of interest for this work
cell_types = cell_types[cell_types['CellType'].isin(cell_types_list)] # parsing sampele type list, to only contain the cell types of interest
cell_types.reset_index(drop=True, inplace=True)
```

```python
#iterating the columns to find sample type for T01 Replicate
counter = 0
for index, row in cell_types.iterrows():
    for col_name in e7_t01_rawCount_data.columns: 
        index_no = e7_t01_rawCount_data.columns.get_loc(col_name)
        if col_name == row["CellID"]:
            counter +=1
            e7_t01_rawCount_data.iat[0,index_no] = row['CellType']

print("Matched Cell Count: " + str(counter))
e7_t01_rawCount_data = e7_t01_rawCount_data.dropna(axis=1)
e7_t01_rawCount_data.to_csv("/Users/krilowcn/Desktop/Projects/GSIS/data/normalizedData/200613_E7_logNorm/updatedFiles/200613_T01_logNorm_updated_cells.csv.gz", index=False, sep=',')
```

```python
#iterating the columns to find sample type for T02 Replicate
counter = 0
for index, row in cell_types.iterrows():
    for col_name in e7_t02_rawCount_data.columns: 
        index_no = e7_t02_rawCount_data.columns.get_loc(col_name)
        if col_name == row["CellID"]:
            counter +=1
            e7_t02_rawCount_data.iat[0,index_no] = row['CellType']

print("Matched Cell Count: " + str(counter))
e7_t02_rawCount_data = e7_t02_rawCount_data.dropna(axis=1)
e7_t02_rawCount_data.to_csv("/Users/krilowcn/Desktop/Projects/GSIS/data/normalizedData/200613_E7_logNorm/updatedFiles/200613_T02_logNorm_updated_cells.csv.gz", index=False, sep=',')
```
